#**🌲🌳🏨 Decision Trees - Hotel Booking**

W oparciu o zbiór [Hotel Booking Cancellation Prediction](https://www.kaggle.com/datasets/youssefaboelwafa/hotel-booking-cancellation-prediction) przeprowadź proces predykcji wykorzystujący drzewa decyzyjne.

```python
import numpy as np
import gdown # pobieranie plików z usługi Google Drive
import pandas as pd # manipulacja i analiza danych
import matplotlib.pyplot as plt # wykresy i wizualizacja danych
from sklearn.model_selection import train_test_split # import funkcji train_test_split do podziału danych na zbiory treningowe i testowe
from sklearn.tree import DecisionTreeRegressor # model regresji oparty o drzewo
from sklearn.ensemble import RandomForestRegressor # model regresji oparty o lasy losowe
from sklearn.tree import plot_tree # funkcja z biblioteki sklearn.tree w Pythonie, która służy do wizualizacji drzewa decyzyjnego
from sklearn.metrics import r2_score # import funkcji r2_score,
                                     # która oblicza współczynnik determinancji
import seaborn as sns
```

```python
#
# kod
dataframe = pd.read_csv("booking.csv")
dataframe
#
```

```python
X = dataframe.drop(columns=['Booking_ID', 'type of meal', 'room type', 'market segment type', 'date of reservation', 'booking status'], axis=1)
y = dataframe['lead time']

X_treningowe, X_testowe, y_treningowe, y_testowe = train_test_split(X, y, test_size=0.2, random_state=0)

drzewo2 = DecisionTreeRegressor(max_depth=5, random_state=0)
drzewo2.fit(X_treningowe, y_treningowe)

predykcja_drzewo2 = drzewo2.predict(X_testowe)
```

```python
# Obliczamy współczynnik determinacji R²
r2_drzewo = r2_score(y_testowe, predykcja_drzewo2)

print(f"Współczynnik determinacji R² dla drzewa decyzyjnego: {r2_drzewo}")
```
