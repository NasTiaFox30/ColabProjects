# Generation and Visualization of Synthetic Economic Data

Zao偶enia dotyczce danych
W obu przykadach symulujemy proste dane ekonomiczne:

* Inflacja (wartoci w procentach)
* Bezrobocie (wartoci w procentach)
* Wzrost PKB (wartoci w procentach)

Zakadamy, 偶e dane te dotycz kolejnych 12 miesicy (od 1 do 12). Oczywicie w rzeczywistych zastosowaniach mo偶na wczyta dane np. z pliku CSV (pochodzcego z Kaggle lub innego 藕r贸da) i analogicznie je zwizualizowa.

```python
# 1. Trzy wykresy obok siebie (wsp贸lna o y)

import matplotlib.pyplot as plt
import numpy as np

# Przykadowe dane (12 miesicy)
miesiace = np.arange(1, 13)
inflacja = [3.1, 3.2, 3.2, 3.5, 3.7, 4.0, 4.2, 4.5, 4.6, 4.4, 4.3, 4.2]
bezrobocie = [5.2, 5.1, 5.1, 5.0, 4.9, 4.8, 4.8, 4.7, 4.7, 4.6, 4.5, 4.5]
wzrost_pkb = [1.2, 1.3, 1.1, 1.4, 1.6, 1.5, 1.7, 1.8, 1.9, 2.0, 2.2, 2.3]

# Utworzenie przestrzeni na wykres i trzech subplots (1 wiersz, 3 kolumny)
fig, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

# Wykres inflacji
axes[0].plot(miesiace, inflacja, marker='o', color='red')
axes[0].set_title("Inflacja")
axes[0].set_xlabel("Miesic")
axes[0].set_ylabel("Procent")

# Wykres bezrobocia
axes[1].plot(miesiace, bezrobocie, marker='o', color='blue')
axes[1].set_title("Bezrobocie")
axes[1].set_xlabel("Miesic")
# O y jest wsp贸dzielona, wic nie musimy ustawia ponownie etykiety

# Wykres wzrostu PKB
axes[2].plot(miesiace, wzrost_pkb, marker='o', color='green')
axes[2].set_title("Wzrost PKB")
axes[2].set_xlabel("Miesic")

# Dopasowanie odstp贸w midzy wykresami i wywietlenie
plt.tight_layout()
plt.show()
```

Wyjanienie kluczowych parametr贸w:

* subplots(1, 3, sharey=True) tworzy 3 wykresy w jednym wierszu, wsp贸dzielce o y (czyli zakres wartoci na osi pionowej bdzie ten sam).
* figsize=(12, 4) kontroluje rozmiar caej figury w calach (szeroko x wysoko).
* plt.tight_layout() dopasowuje automatycznie odstpy midzy wykresami, by nie nachodziy na siebie etykiety.
