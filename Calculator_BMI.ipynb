Zacznijmy od funkcji liczƒÖcej wska≈∫nik masy cia≈Ça, czyli Body Mass Index (BMI).

Jak widzisz, formu≈Ça przyjmuje dwie warto≈õci:

wagƒô (pierwotnie w kilogramach)
wzrost (pierwotnie w metrach)
Napisz funkcjƒô bmi, kt√≥ra bƒôdzie przyjmowaƒá dwa argumenty.

Napisz program, kt√≥ry pobiera od u≈ºytkownika wagƒô w kilogramach oraz wzrost w metrach, nastƒôpnie za pomocƒÖ funkcji bmi oblicza wska≈∫nik i informuje u≈ºytkownika czy u≈ºytkownik ma niedowagƒô, prawid≈ÇowƒÖ wagƒô lub nadwagƒô.

```python
# Kalkulator BMI
waga = input("Podaj wagƒô w kg - ")
wzrost = input("Podaj wzrost w m - ")

def obliczBMI(waga, wzrost):
  bmi = float(waga) / (float(wzrost)**2)
  wskaznik = round(bmi,2)
  print("Tw√≥j BMI = ",bmi)
  if (bmi < 16.0 ):
    print("wyg≈Çodzenie")
  elif (16 < bmi < 16.99 ):
    print("wychudzenie")
  elif (17.0 < bmi < 18.49 ):
    print("niedowaga")
  elif (18.5 < bmi < 24.99 ):
    print("po≈ºƒÖdana masa cia≈Ça")
  elif (25.0 < bmi < 29.99 ):
    print("nadwaga")
  elif (30.0 < bmi < 34.99 ):
    print("oty≈Ço≈õƒá I stopnia")
  elif (35.0 < bmi < 39.99 ):
    print("oty≈Ço≈õƒá II stopnia (du≈ºa)")
  elif (bmi > 40.0 ):
    print("oty≈Ço≈õƒá III stopnia (chorobliwa)")

obliczBMI(waga, wzrost)
```
