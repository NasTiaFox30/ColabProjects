**Zadanie 2**:

StwÃ³rz histogram (distplot) z uÅ¼yciem danych z zestawu danych tips. Wykorzystaj kolumnÄ™ total_bill jako wartoÅ›ci do utworzenia histogramu.

RozwiÄ…zanie:

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.distplot(tips['total_bill'], bins=10)
plt.show()
```

**Zadanie 3**:

StwÃ³rz heatmap (mapa ciepÅ‚a) z uÅ¼yciem danych z zestawu danych flights. Wykorzystaj kolumnÄ™ month jako oÅ› X, a kolumnÄ™ year jako oÅ› Y, a wartoÅ›ciÄ… w komÃ³rce niech bÄ™dzie liczba pasaÅ¼erÃ³w.

RozwiÄ…zanie:

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')
flights = flights.pivot(index='month',columns='year', values='passengers')

sns.heatmap(flights)
plt.show()
```

Zadanie 4:

StwÃ³rz wykres punktowy (scatter plot) z uÅ¼yciem danych z zestawu danych tips. Wykorzystaj kolumnÄ™ total_bill jako oÅ› X, a kolumnÄ™ tip jako oÅ› Y. Dodaj dodatkowy wymiar koloru do wykresu, ktÃ³ry odpowiada kolumnie sex.

RozwiÄ…zanie:

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.show()
```

**Zadanie 5**:

StwÃ³rz wykres liniowy (line plot) z uÅ¼yciem danych z zestawu danych fmri. Wykorzystaj kolumnÄ™ timepoint jako oÅ› X, a kolumnÄ™ signal jako oÅ› Y. Dodaj dodatkowy wymiar koloru do wykresu, ktÃ³ry odpowiada kolumnie region.

RozwiÄ…zanie:

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset('fmri')

sns.lineplot(x='timepoint', y='signal', hue='region', data=fmri)
plt.show()
```
