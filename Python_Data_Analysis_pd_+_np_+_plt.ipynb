```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/plotly/datasets/master/nz_weather.csv'

df = pd.read_csv(url)
```

## Wypisz piƒôƒá pierwszych wynik√≥w

```python
# Tw√≥j kod
print(df.head(10))
```

## Narysuj dwie linie na jednym wykresie przedstawiajƒÖce pogodƒô w Christchurch i Hamilton w ciƒÖgu roku

```python
plt.figure(figsize=(10,8))
x = df.DATE
y_1 = df.Christchurch
y_2 = df.Hamilton

plt.plot(x,y_1, label = "Christchurch")
plt.plot(x,y_2, label = "Hamilton")

plt.xticks(np.arange(0, 150, 8))
plt.xticks (rotation=30)
plt.yticks(np.arange(0, 150, 5))

plt.legend()
plt.show()
```

## Stw√≥rz histogram dla pogody w Auckland

```python
y = df.Auckland

plt.hist(y,bins=50, color='green')

plt.show()
```

## Stw√≥rz wykres pude≈Çkowy dla pogody w Auckland

```python
y=df.Auckland

plt.boxplot(y)
plt.show()
```

## Stw√≥rz podwykres przedstawiajƒÖcy wszystkie wykresy pude≈Çkowe dla poszczeg√≥lnych obszar√≥w

```python
# Auckland Christchurch Dunedin Hamilton Wellington

y_1 = pd.to_numeric(df.Auckland, errors='coerce')
y_2 = pd.to_numeric(df.Christchurch, errors='coerce')
y_3 = pd.to_numeric(df.Dunedin, errors='coerce')
y_4 = pd.to_numeric(df.Hamilton, errors='coerce')
y_5 = pd.to_numeric(df.Wellington, errors='coerce')

fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1, 5, sharey=True)
fig.suptitle('Pogoda w regionach')
# ax1.boxplot(y_1)
ax1.boxplot(y_1.dropna(), patch_artist=True)
ax2.boxplot(y_2.dropna(), patch_artist=True)
ax3.boxplot(y_3.dropna(), patch_artist=True)
ax4.boxplot(y_4.dropna(), patch_artist=True)
ax5.boxplot(y_5.dropna(), patch_artist=True)


plt.show()
```
